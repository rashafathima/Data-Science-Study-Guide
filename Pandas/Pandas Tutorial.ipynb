### ● Joins

```python
# Creating a table

data_sm1 = pd.DataFrame({  #Data frame is a datatype that defines the datasets is of the form table 
    "Student" : ["Akash","Joe","Sony"],
    "Age" :["15","17","8"]
})

data_sm1
```

```python
data_sm2 = pd.DataFrame({
    "Student" : ["Akash","Joe","Muna"],
    "Marks" : ["80","45","98"]
})

data_sm2
```

```python
# Inner Join

pd.merge(data_sm1,data_sm2,on="Student") #inner join by default

# Inner join finds the similarity in both the datasets . Here we have set the condition that we would be merging the datasets depending on the student name . If the student name on both the datsets are same we will merge them or else we will drop them
```

```python
#Outer Join

pd.merge(data_sm1,data_sm2,on="Student", how="outer") 

#Outer join merges all the data in both the datasets and displays them
```

```python
# Left Join

# Considers all the data on the left sided dataset (here left sided datset is data_sm1) and only the data in the dataset on the right side having similarity with the left sided datset
pd.merge(data_sm1,data_sm2,on="Student", how="left") 
```

```python
# Right Join

# Considers all the data on the right sided dataset (here right sided datset is data_sm2) and only the data in the dataset on the left side having similarity with the right sided datset
pd.merge(data_sm1,data_sm2,on="Student", how="right") 
```

### ● Pivoting

```python
piv_sam = pd.read_csv('C:/Users/Rasha/Downloads/pivot_sample.csv')
piv_sam
```

```python
# Pivot the data

piv_sam.pivot(index ="Page",columns="Location") #Selecting index means selecting the classes of whose brief data you want to display

#The o/p will show the hits of a particular page at different location (concentrated attribute : Page)
```

```python
#Pivot the table

piv_sam.pivot_table(index = "Page", aggfunc = "sum")

#finds and calculates the total number of hits for every page (ignored attribute = Location)
```

```python
piv_mean_loc = piv_sam.pivot_table(index = "Location", aggfunc = "mean")

#finds the mean of total number of hits for every location (ignored attribute = Page)
piv_mean_loc

#for later reference ive created this variable piv_mean_loc
```

```python
piv_sam.pivot_table(index = "Location", aggfunc="count") #no ignored attribute 
```

```python
#To find the grand total of the column 'hits'

piv_sam['Hits'].sum()
```

```python
# Plotting a bar chart that describes the number of hits at different locations

piv_mean_loc.plot(kind = 'bar')

#note : the piv_mean_loc describes the mean of total hits at locations GOA, MUMBAI AND HAWAII
```

```python
#Line chart of total hits vs the index(0-11) 

piv_sam.plot(kind = 'line')
```

### ● Shifting

```python
# Shifting the row by one position down 

fb_stock = pd.read_csv('C:/Users/Rasha/Downloads/FB.csv')
fb_stock.head()

fb_stock.shift(1).head()
```

```python
# Shifting the row by one position above

fb_stock.shift(-1).head()
```

```python
#Calculating the prvious closing price.(for the 2nd row take the value of 1st row close column)

fb_stock['prev_close'] = fb_stock['Close'].shift(1)

fb_stock.head()
```

```python
# To build a column that depicts the diff bw close column values and the prev_close

fb_stock['diff_close'] = fb_stock['Close']-fb_stock['prev_close']

fb_stock.head()
```

```python
# Generating Weekly returns

fb_stock['weeky_returns'] = ((fb_stock['Close']-fb_stock['prev_close'].shift(7))/fb_stock['Close'])*100
# Generates the weekly returns hence the first seven rows are empty(shift 7) as theres no prev week before the first week. Then i calculate the percentage of weekly returns
fb_stock.head()
```

```python
# Writing to an excel sheeft/csv file. (i.e exporting the csv file from notebook to drive)
# Here our data is stored in varible fb_stock

fb_stock.to_csv('C://Users/Rasha/Downloads/fb_exported.csv' #you can also export the data having specific columns by mentioning columns = ['Open','High']. example)
```
